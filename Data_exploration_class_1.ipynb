## Why You Should Use Virtual Environments

### What is a Virtual Environment?

A **virtual environment** is an isolated Python workspace on your computer. Think of it like a separate folder that contains:
- Its own Python interpreter
- Its own set of installed packages
- Its own package versions

Instead of installing all packages globally (across your entire system), you create separate environments for different projects.

### Problems With Global Python Installations

**❌ WITHOUT Virtual Environments (Working Globally):**

1. **Version Conflicts**: You have multiple projects that need different versions of the same package
   - Project A needs `pandas 1.0`
   - Project B needs `pandas 2.1`
   - You can only have ONE version installed globally → **Projects break**

2. **Dependency Hell**: Installing a package for one project accidentally updates another package globally
   - You install `scikit-learn`, which upgrades `NumPy`
   - Your old project relied on the old `NumPy` version → **Broken code**

3. **System Pollution**: All packages accumulate in one place
   - Your Python folder becomes cluttered with hundreds of packages
   - Hard to know which packages each project actually needs
   - Harder to share projects with others

4. **Reproducibility Issues**: Someone else runs your code with different package versions
   - Your code works on your computer but **fails on theirs**
   - No clear record of which versions you used

5. **Difficult Cleanup**: Removing packages you don't need is risky
   - You might break something else that depends on it
   - Can't safely uninstall without breaking projects

### Benefits of Virtual Environments

**✅ WITH Virtual Environments (Best Practice):**

| Benefit | Explanation |
|---------|-------------|
| **Isolation** | Each project has its own packages. No cross-project conflicts. |
| **Version Control** | Project A can use `pandas 1.0`, Project B uses `pandas 2.1`. Both work perfectly. |
| **Clean System** | Global Python stays clean. Only the packages YOU need are installed. |
| **Easy Sharing** | Create `requirements.txt` → others clone your environment exactly → code works first try. |
| **Safe Cleanup** | Delete a virtual environment = completely remove all its packages. No side effects. |
| **Reproducibility** | Anyone can recreate your exact environment. Your code runs on all computers. |
| **Professional Standard** | All professional Python developers use virtual environments. Industry best practice. |

```bash
# Deactivate when done

```

## Understanding the Diamonds Dataset: What You're Researching

Before diving into analysis, it's crucial to understand **what each variable represents** and **why it matters**. Here's a complete guide to the 10-column diamonds dataset:

### **The 4Cs of Diamond Quality** (The Core Quality Metrics)

#### **1. `carat`** (continuous, numeric)
- **What it is**: A unit of weight for diamonds. 1 carat = 200 milligrams.
- **Why it matters**: 
  - Heavier diamonds are typically more valuable and rare
  - Non-linear relationship with price (doubling weight doesn't double price)
  - Strong predictor of price
- **Data used**: 0.2 to 5.01 carats
- **Practical note**: Price per carat increases exponentially above 1 carat (psychological pricing threshold)

#### **2. `cut`** (categorical, ordinal)
- **What it is**: The quality of the diamond cut — how well the diamond is proportioned and polished
- **Categories**: Fair → Good → Very Good → Premium → Ideal
- **Why it matters**:
  - Affects how much light the diamond reflects (brilliance)
  - Better cuts command higher prices for same weight/clarity
  - Ordinal: there's a natural ranking order
- **Practical note**: A well-cut diamond sparkles more; a poor cut wastes light even in quality stones

#### **3. `color`** (categorical, ordinal)
- **What it is**: The absence of color in the diamond (colorless diamonds are most valuable)
- **Scale**: D, E, F (colorless) → G, H, I, J (near-colorless) → K-Z (faint color)
- **Why it matters**:
  - Most people cannot distinguish between D and G with the naked eye
  - Colorless diamonds (D-F) command premium prices
  - Creates natural price tiers
- **Practical note**: Diamonds below H become visibly yellowed to trained eyes

#### **4. `clarity`** (categorical, ordinal)
- **What it is**: The absence of inclusions (internal defects) and blemishes (surface marks)
- **Scale**: I1 (worst) → SI2, SI1 → VS2, VS1 → VVS2, VVS1 → IF, FL (flawless)
- **Why it matters**:
  - Inclusions can weaken diamond and reduce transparency
  - Only VVS and above are "eye-clean" to most people under normal viewing
  - Creates discrete price categories
- **Practical note**: Most engagement diamonds are VS1/VS2 (good clarity at reasonable cost)

---

### **Physical Dimensions** (The Measurements)

#### **5. `depth`** (continuous, numeric)
- **What it is**: The total depth of the diamond from top to bottom, as a percentage: `depth% = (total depth) / (average diameter) × 100`
- **Typical range**: 43% to 79%
- **Why it matters**:
  - Indicates if diamond is well-proportioned or poorly cut
  - Ideal depth typically 59-63% (too deep or too shallow = light wasted/lost)
  - Inverse relationship with price for given carat (deeper = material wasted)
- **Practical note**: Used to assess cut quality; works with `table` to diagnose poor proportions

#### **6. `table`** (continuous, numeric)
- **What it is**: The width of the diamond's top facet as a percentage of the overall diameter
- **Typical range**: 43% to 95%
- **Why it matters**:
  - Table determines the "window" through which light enters
  - Too small (narrow): not enough light enters
  - Too large (wide): light scatters inefficiently
  - Ideal range 54-57% for round diamonds
- **Practical note**: Extreme values (very high or very low) predict poor cut quality and lower prices

#### **7. `x`, `y`, `z`** (continuous, numeric)
- **What they are**: Physical dimensions in millimeters
  - **`x`** = length (mm)
  - **`y`** = width (mm)
  - **`z`** = depth (mm)
- **Why they matter**:
  - Direct physical measurements; combine to approximate volume
  - Relationship to carat: heavier diamonds are larger (but cut affects proportions)
  - Extreme values or unusual proportions (`x ≠ y ≠ z`) flag measurement errors or unusual shapes
  - Outliers here can cause huge prediction errors when back-transforming from log scale
- **Practical note**: If x, y, z are too extreme, the diamond may be mislabeled or data entry error; create verification rule that x ≈ y ≈ z

---

### **The Target Variable**

#### **8. `price`** (continuous, numeric) ⭐ *This is what we predict*
- **What it is**: The price of the diamond in USD
- **Typical range**: $326 to $18,823
- **Why it matters**:
  - Our prediction target
  - Heavily right-skewed (most cheap, few expensive)
  - Log-transformation recommended to reduce skewness
  - Non-negative by definition
- **Practical note**: Real diamonds used in pricing studies; data sourced from dealers — not all price range equally represented

---

### **Why Knowing Your Data Matters: Industry Context**

The diamonds dataset is **real marketplace data**, which means:

1. **The 4Cs are not independent**
   - Larger diamonds (higher carat) are rarer in high clarity/color
   - Affects how coefficients behave in regression

2. **Price is heavily influenced by psychological thresholds**
   - Specific carat weights (0.5, 1.0, 1.5, 2.0) command price premiums
   - Color grades (D-F vs G-J) create discrete pricing tiers

3. **Outliers have domain explanations**

4. **What we're NOT looking at** (assumes quality is constant)
   - Fluorescence (glows under UV)
   - Grading lab (GIA, AGS, EGL use different standards)
   - Certification status
   - Geographic origin (impacts rarity/value)

## Fast-track block: outlier finding (45-minute version)

This compact block uses a simple SD-rule to flag potential outliers in key numeric predictors and visualize where they appear against `price`.

````python
dim_cols = [c for c in ["x", "y", "z"] if c in df_work.columns]
zero_dim_mask = pd.Series(False, index=df_work.index)

if dim_cols:
    zero_dim_mask = (df_work[dim_cols] == 0).any(axis=1)

n_zero_dim = int(zero_dim_mask.sum())
if n_zero_dim > 0:
    print(f"Found {n_zero_dim} rows where at least one of x/y/z is zero. Dropping these rows as likely data errors.")
    df_work = df_work.loc[~zero_dim_mask].copy()
else:
    print("No rows found with x/y/z equal to zero.")

key_vars = [v for v in ["table", "x", "y", "z"] if v in df_work.columns]
outlier_rows = []

for v in key_vars:
    mu = df_work[v].mean()
    sd = df_work[v].std()
    flag = (df_work[v] < mu - 3.5 * sd) | (df_work[v] > mu + 3.5 * sd)
    outlier_rows.append({
        "variable": v,
        "mean": mu,
        "std": sd,
        "outlier_count": int(flag.sum()),
        "outlier_pct": float(flag.mean() * 100),
    })

outlier_summary_45 = pd.DataFrame(outlier_rows).sort_values("outlier_pct", ascending=False)
display(outlier_summary_45.round(3))

if key_vars:
    fig, axes = plt.subplots(1, len(key_vars), figsize=(6 * len(key_vars), 4), squeeze=False)
    for i, v in enumerate(key_vars):
        mu = df_work[v].mean()
        sd = df_work[v].std()
        flag = (df_work[v] < mu - 3.5 * sd) | (df_work[v] > mu + 3.5 * sd)
        axes[0, i].scatter(df_work.loc[~flag, v], df_work.loc[~flag, "price"], alpha=0.25, s=12, label="inlier")
        axes[0, i].scatter(df_work.loc[flag, v], df_work.loc[flag, "price"], alpha=0.8, s=16, color="tomato", label="outlier")
        axes[0, i].set_title(f"{v} vs price")
        axes[0, i].set_xlabel(v)
        axes[0, i].set_ylabel("price")
    axes[0, 0].legend(loc="best")
    plt.tight_layout()
    plt.show()
````

### Quick interpretation of outlier panel

- Focus on **rate**, not only count: a few extreme points can still strongly affect fitted lines.
- Rows where any of `x`, `y`, `z` were exactly zero were removed first as likely data errors.
- The SD rule is a screening tool, not automatic deletion for the remaining non-zero rows.
- Continue with modeling after this basic data-quality cleanup.

## Correlation Analysis: Spearman vs Pearson

````python
from scipy.stats import spearmanr, pearsonr

# Select numeric columns for correlation analysis
numeric_subset = df_work[[c for c in df_work.columns if c in ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']]].copy()

# Compute Pearson and Spearman correlations with price
corr_rows_comparison = []
for col in numeric_subset.columns:
    if col != 'price':
        try:
            pearson_coef, pearson_p = pearsonr(numeric_subset[[col, 'price']].dropna().iloc[:, 0], 
                                               numeric_subset[[col, 'price']].dropna().iloc[:, 1])
            spearman_coef, spearman_p = spearmanr(numeric_subset[[col, 'price']].dropna().iloc[:, 0], 
                                                  numeric_subset[[col, 'price']].dropna().iloc[:, 1])
            corr_rows_comparison.append({
                'feature': col,
                'pearson_r': pearson_coef,
                'pearson_p': pearson_p,
                'spearman_rho': spearman_coef,
                'spearman_p': spearman_p,
                'diff_abs': abs(pearson_coef - spearman_coef),
            })
        except:
            pass

corr_summary = pd.DataFrame(corr_rows_comparison).sort_values('diff_abs', ascending=False)
print("=== Pearson vs Spearman Correlations with Price ===")
display(corr_summary.round(4))

# Plot comparison
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Pearson
axes[0].barh(corr_summary['feature'], corr_summary['pearson_r'], color='steelblue', alpha=0.7)
axes[0].set_xlabel('Pearson r')
axes[0].set_title('Pearson Correlation with Price')
axes[0].axvline(0, color='black', linestyle='--', lw=0.8)

# Spearman
axes[1].barh(corr_summary['feature'], corr_summary['spearman_rho'], color='coral', alpha=0.7)
axes[1].set_xlabel('Spearman ρ')
axes[1].set_title('Spearman Correlation with Price')
axes[1].axvline(0, color='black', linestyle='--', lw=0.8)

plt.tight_layout()
plt.show()

# Highlight differences
print("\n=== Interpretive Notes ===")
print("Pearson measures LINEAR relationship; Spearman measures MONOTONIC relationship.")
print("Large differences suggest nonlinear monotonic patterns (outliers/rank-driven effects).")
print(f"Max |r - ρ| = {corr_summary['diff_abs'].max():.4f} ({corr_summary.iloc[0]['feature']})")
````

### How to interpret Pearson vs Spearman here

- **All key size variables (`carat`, `x`, `y`, `z`) show very strong positive association with price** under both measures.
- **Spearman is consistently higher than Pearson** for `x`, `y`, `z`, and `carat`. This usually indicates a strong **monotonic but not perfectly linear** relationship (and/or sensitivity of Pearson to tail behavior).
- The largest gaps are for **`y` and `z`** (`diff_abs` around 0.09 to 0.10), suggesting stronger rank-order agreement than linear fit on the raw scale.
- **`table`** has a weak positive relationship with price (small but significant), so it contributes less than size/weight variables.
- **`depth`** is near zero in both metrics, meaning little standalone marginal association with price in this sample.

Practical modeling takeaway:

- For prediction, keep `carat` and dimensions as core drivers.
- Because some relationships are not strictly linear on the raw scale, transformations and robust diagnostics are justified.
- Statistical significance is expected with this sample size, so focus on **effect size** (`r`, `rho`) and practical importance, not only p-values.

## Regression Modeling: Prediction & Statistical Inference

### What is One-Hot Encoding (Categorical Dummies)?

**The Problem:**
Regression models require numeric inputs. But we have categorical features like cut="Ideal", clarity="VS1".

**The Solution: One-Hot Encoding (Dummies)**
Convert each categorical value into binary (0/1) indicators. Example:

```
Original: cut="Fair", "Good", "Very Good", "Premium", "Ideal"
            ↓ One-Hot Encode ↓
cut_Fair=1, cut_Good=0, cut_VeryGood=0, cut_Premium=0, cut_Ideal=0
cut_Fair=0, cut_Good=1, cut_VeryGood=0, cut_Premium=0, cut_Ideal=0
... etc
```

**Key Principle: drop_first=True**
When you have n categories, create n-1 dummies. (Omit one category to avoid multicollinearity.)
- Why? If all 5 cut dummies exist AND an intercept, they're linearly dependent (perfect multicollinearity).
- Solution: Drop one level (e.g., cut_Fair) as the baseline reference. All other dummy coefficients report the *difference* from that baseline.

**How to Interpret:**
- `cut_Good_coef = +0.15` → "Good" diamonds are 0.15 log-price units HIGHER than Fair (baseline)
- `clarity_I1_coef = -0.50` → Inclusion class I1 is 0.50 log-price units LOWER than the baseline clarity

This is implemented via `pd.get_dummies(X, drop_first=True)` in pandas and `sklearn.preprocessing.OneHotEncoder`.


````python
# Identify categorical columns in X
cat_cols_in_X = X.select_dtypes(include=['object', 'category']).columns.tolist()

# Create one-hot encoded dummies, dropping the first category to avoid multicollinearity
X = pd.get_dummies(X, columns=cat_cols_in_X, drop_first=True, dtype=int)

print(f"Shape after adding dummies: {X.shape}")
print(f"New columns:\n{X.columns.tolist()}")
````

````python
import statsmodels.api as sm
import pandas as pd

# Use cleaned working dataframe
df_reg_use = df_work.dropna().copy()

y = df_reg_use["price"]
X = df_reg_use.drop(columns=["price","log_price"])

# Ensure that all columns in X are numeric (convert bool to int)
X = pd.get_dummies(X, drop_first=True)
X = X.astype(int)

# Add constant term (intercept) to the predictors
X = sm.add_constant(X)

# Fit the model
model = sm.OLS(y, X).fit()

# Print the regression summary (coefficients, p-values, etc.)
print(model.summary())
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import numpy as np

# Plot the residuals
fig, ax = plt.subplots(figsize=(10, 5))
sns.residplot(x=model.fittedvalues, y=model.resid, lowess=True, line_kws={'color': 'red'})
ax.set_title("Residuals Plot")
plt.show()

# Check for residuals' histogram to see their distribution
fig, ax = plt.subplots(figsize=(10, 5))
sns.histplot(model.resid, kde=True, bins=30, color='blue')
ax.set_title("Residuals Distribution")
plt.show()

# Plot Actual vs Predicted (True vs Predicted y)
fig, ax = plt.subplots(figsize=(10, 5))
ax.scatter(y, model.fittedvalues, alpha=0.6, color='green')
ax.plot([y.min(), y.max()], [y.min(), y.max()], '--', color='red')  # Add 45-degree line
ax.set_xlabel('Actual y')
ax.set_ylabel('Predicted y')
ax.set_title('Actual vs Predicted')
plt.show()

# QQ-Plot of the residuals
fig, ax = plt.subplots(figsize=(10, 5))
sm.qqplot(model.resid, line='45', fit=True, ax=ax)
ax.set_title('QQ Plot of Residuals')
plt.show()

# ==========================================
# REMOVING OUTLIERS & REFITTING
# ==========================================
print("\n--- Refitting model after removing outliers ---")

# Identify outliers using Cook's distance
influence = model.get_influence()
cooks_distances = influence.cooks_distance[0]
n = len(y)
# Common threshold for Cook's distance is 4 / n
outlier_threshold = 4 / n
outlier_mask = cooks_distances < outlier_threshold

X_clean = X[outlier_mask]
y_clean = y[outlier_mask]

print(f"Removed {(~outlier_mask).sum()} outliers based on Cook's Distance.")

# Refit the model
model_clean = sm.OLS(y_clean, X_clean).fit()

# Re-plot Residuals
fig, ax = plt.subplots(figsize=(10, 5))
sns.residplot(x=model_clean.fittedvalues, y=model_clean.resid, lowess=True, line_kws={'color': 'red'})
ax.set_title("Residuals Plot (No Outliers)")
plt.show()

# Re-plot QQ-Plot
fig, ax = plt.subplots(figsize=(10, 5))
sm.qqplot(model_clean.resid, line='45', fit=True, ax=ax)
ax.set_title('QQ Plot of Residuals (No Outliers)')
plt.show()
````

````python
# ==========================================
# REMOVING OUTLIERS BASED ON X DATA (Features)
# ==========================================
print("\n--- Refitting model after removing outliers in X data ---")

# Let's identify the continuous numeric columns in X to check for outliers
# We ignore the constant ('const') and any binary dummy variables
continuous_cols = [c for c in X.columns if c != 'const' and X[c].nunique() > 2]

# Calculate Z-scores for these continuous columns
z_scores = np.abs((X[continuous_cols] - X[continuous_cols].mean()) / X[continuous_cols].std())

# Keep rows where all Z-scores are less than 3 (the standard threshold for outliers)
outlier_mask_data = (z_scores < 3).all(axis=1)

X_clean_data = X[outlier_mask_data]
y_clean_data = y[outlier_mask_data]

print(f"Original dataset size: {len(X)}")
print(f"Removed {(~outlier_mask_data).sum()} outliers based on feature Z-scores > 3.")
print(f"New dataset size: {len(X_clean_data)}\n")

# Refit the model on the data-cleaned sets
model_clean_data = sm.OLS(y_clean_data, X_clean_data).fit()

# Plot Actual vs Predicted
fig, ax = plt.subplots(figsize=(10, 5))
ax.scatter(y_clean_data, model_clean_data.fittedvalues, alpha=0.6, color='green')
ax.plot([y_clean_data.min(), y_clean_data.max()], [y_clean_data.min(), y_clean_data.max()], '--', color='red')
ax.set_xlabel('Actual y')
ax.set_ylabel('Predicted y')
ax.set_title('Actual vs Predicted (No Data Outliers)')
plt.show()

# Re-plot Residuals
fig, ax = plt.subplots(figsize=(10, 5))
sns.residplot(x=model_clean_data.fittedvalues, y=model_clean_data.resid, lowess=True, line_kws={'color': 'red'})
ax.set_title("Residuals Plot (No Data Outliers)")
plt.show()

# Re-plot QQ-Plot
fig, ax = plt.subplots(figsize=(10, 5))
sm.qqplot(model_clean_data.resid, line='45', fit=True, ax=ax)
ax.set_title('QQ Plot of Residuals (No Data Outliers)')
plt.show()

print(model_clean_data.summary())
````
